##3. Visualize data in a Map

```python
# here a map will be created
m = folium.Map(location=[47.80441683656554, 13.04870460299294], zoom_start=12, control_scale=True)
```

```python
# add markers
folium.Marker(
    [47.80405845892574, 13.037556651700381], popup = "Calisthenics <i>Salzach</i>", tooltip = "Calisthenics",
    icon = folium.Icon(color = "red", icon="dumbbell", prefix='fa')
).add_to(m)

folium.Marker(
    [47.777922927461255, 13.064586591558372], popup = "Calisthenics <i>Alpenstraße</i>", tooltip = "Calisthenics",
    icon = folium.Icon(color = "red", icon="dumbbell", prefix='fa')
).add_to(m)
m

# normally I would create a loop over each sport sites of the same sport category. Each sport category could get a unique icon. 
# This is more for demonstration how it works with folium.
```

Like that you can set different markers and also choose set different tooltips and popups and also choose different icons (see https://fontawesome.com/icons)

```python
# adding a different base map (imagery base map)
tile = folium.TileLayer(
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='Esri',
    name='Esri Satellite',
    overlay=False,
    control=True
).add_to(m)

# allowing to switch between osm base map and esri satelite imagery base map
folium.LayerControl().add_to(m)

# show map
m
```
